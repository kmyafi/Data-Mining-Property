## *Handling outlier*

Akan dilakukan pemisahan terhadap outlier dari data

### Harga

```python
sns.boxplot(df_full['harga'], orient="h")
```

```python
# IQR
Q1 = np.percentile(df_full['harga'], 25, method='midpoint')
Q3 = np.percentile(df_full['harga'], 75, method='midpoint')
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

# Above Upper bound
upper_array = np.array(df_full['harga'] > upper)
print("Upper Bound:", upper)
print(upper_array.sum())

#Below Lower bound
lower_array = np.array(df_full['harga'] < lower)
print("Lower Bound:", lower)
print(lower_array.sum())
```

```python
outliers = df_full[(df_full['harga'] < lower) | (df_full['harga'] > upper)]
outliers
```

```python
# Separating outliers from the full data
df_full.drop(outliers.index, inplace = True)
```

6. Jika RPPI ingin melakukan investasi (max) Rp. 25 Milyar minggu besok, apakah dari data anda dapat memberikan rekomendasi rumah mana saja yang akan berpotensi menghasilkan keuntungan bagi perusahaan?

```python
affordable_properties = df_full[df_full['harga'] <= 25000000000]

# Profit Calculation
affordable_properties['potential_profit'] = 25000000000 - affordable_properties['harga']
sorted_properties = affordable_properties.sort_values(by='potential_profit', ascending=False)

# recommended properties
print("\nTop Recommended Properties for Investment:")
print(sorted_properties[['harga', 'potential_profit']].head(10))
```

Logic no. 6: Dicari rumah yang di bawah 25B, agar *investment* tidak hanya di satu properti (*maximize capital*)/diversitas portfolio
